## 摘要

&emsp;&emsp;本研究旨在探究个体的时间使用碎片化程度与个体自我感知的快乐程度与焦虑程度之间的关系。本研究采用python语言作为分析工具，以英国时间使用调查数据为材料，使用香农熵——时间是用碎片化、设备数、自我报告的快乐程度、自我报告的焦虑程度为变量，对4360份样本数据进行定量分析。结果表明：（1）性别与年龄会影响时间使用碎片化的程度。女性和年长的群体在时间使用上碎片化程度更高。（2）个体拥有的电子设备的数量越多，时间使用碎片化的程度越高。（3）时间使用碎片化越高，个体自我感知的快乐程度越低，焦虑程度越高。

## 一、文献综述

&emsp;&emsp;我们的日常生活是被时间组织起来的，如吉登斯所言：“即使是最复杂的社会组织形式，从根本上说也是日常生活的惯例常规构成的。”从日出而作日落而息的前现代社会，到依照钟表时间安排生活事务的工业化时代，再到信息革命之后的当下，去时间化开始成为一种新的社会组织模式。卡斯特在《网络社会的崛起》一书中详细介绍了“无时间的时间”以及与之相伴的“流动空间”概念的内涵。其中，无时间之时间是指按照时间密集排列或根据顺序的瞬间而即时排列的社会行为的先后顺序，流动空间是经由信息流动形成的特殊的空间形式，这二者共同构成了网络生活的物质基础。卡斯特用这一组概念精准地描述了信息化社会下由通信技术变革所引发的时空变迁与杜会结构之间的关系，在这样的社会结构之下，工作方式的弹性、不确定化与劳动的个人化成为人们日常生活更为普遍的状态。  

&emsp;&emsp;麦克卢汉指出：“在信息瞬息万里的电子时代，时间（按视觉和切割计算的时间）和空间（统一的、形象的和有周边密封的空间）已不复存在。”我们可以明显感受到，移动互联网的不断发展进一步改造了人们对于时空的认知，在手机普及之后，人们的日常生活不再拘泥于一种特定的在场形式，由于数字媒介已经深度嵌入到人们日常生活的组织秩序之中，过去线性的时间安排被切成无数的碎片，空间上的缺场反而成为了更为普遍的生活状态。这一方面丰富了我们的日常生活体验，但同时又造成了当下时代的萎缩。基于这种矛盾的判断，本文认为人们更加丰富的日常生活似乎并不绝对意味着整体幸福感的提升，甚至可能出现相反的结果。  




### 1.时间碎片化
&emsp;&emsp;通常而言，时间的碎片化意味着生活中零散时间的增多，比如等餐、堵车的时间。但这种理解是经验性的，只描述了碎片时间，而不是整体意义上的碎片化，并且这种描述不能作为一种社会结构的组织方式被操作，因此本文需要对时间的碎片化做出更为准确的定义和操作化。
当今世界知名的社会批判理论家哈特穆特·罗萨（Hartmut Rosa）在他的著作《加速：现代时间结构的改变》以及《新异化的诞生》中详细论述了加速作为现代社会的内在动力如何改造了人们同时间的关系。他认为社会加速异化了人们同时间的关系，使得体验和记忆的时间都开始变得短暂。晚期现代，人们越来越常参与一些非常断裂孤立的活动或脉络，这些活动造成了孤立的行动和体验片段，无法整合地或有意义地将彼此联结在一起。最终，我们没有让我们所体验到的时间变成“我们自己的”时间。  

&emsp;&emsp;这一异化结果在经验层面上被罗萨总结为生活步调的加速。这种加速可以定义为，在一定时间单位当中行动事件量或体验事件量的增加，也就是时间的碎片化。  

&emsp;&emsp;罗萨认为可以从主观和客观两个方面来测量这种生活步调的加速。主观方面，生活速度的加快，可能在个人的时间体验上有可观察到的效应：它会让人们觉得时间很匮乏，觉得很忙碌，感受到紧迫与压力。客观方面，既可以测量可界定出来的行动所耗费的时间区间或“单位”的缩短，比方测量吃饭、睡觉、散布、娱乐、家庭谈心等；也可以测量行动时间与体验时间的“压缩”，像是在一定时间段当中，通过减少休息或间隔时间，而做得更多或体验到更多，或是同时做更多事，也就是“多任务”的时间安排模式。  

&emsp;&emsp;此外，国内现目前针对时间碎片化的研究较多局限于微观的新闻传播领域，关注新闻、营销等文本的碎片化表达。笔者在CNKI上以“时间碎片化”为主题，共检索到4864条中文结果，其中与新闻传播专业相关的结果有990条。但几乎都在讨论具体文本的碎片化传播过程，如碎片化营销、碎片化阅读、以及一些特定知识的碎片化传播的问题，鲜有讨论时间碎片化与日常生活关系的研究。


### 2.时间使用调查

&emsp;&emsp;时间使用调查（Time-use Survey，简称TUS）是一种通过调查区域内特定人群的时间使用情况，以获得各类人群各种活动的时间分布状况的统计学方法。但这种调查模式在中国并没有得到普及，一直到2018年，国家统计局才组织开展了第二次全国时间使用调查。并且，针对时间使用调查结果所做出的研究也较多地被限制在教育、管理等政策领域，对于时间如何组织了人们的日常生活以及如何影响了人们的日常生活质量方面鲜有涉及。  

&emsp;&emsp;关于时间使用调查方面的研究在西方国家更为普及，也更加贴近日常生活。比如José Ignacio Giménez Nadal等人于2019年发布了一篇基于美国时间使用调查结果进行的有关在家工作时间与工人幸福感关系的研究；Song Younghwan于2018年发布一份同样基于美国时间使用调查结果的有关工作流离失所与主观幸福感之间关系的研究；Holmes Erin Kramer等人于2021年发布了一份同样基于美国时间使用调查结果的研究，关注了留守父亲的幸福感问题。  


## 二、研究思路与假设

&emsp;&emsp;根据对过往文献的梳理，可以看出不同的时间使用策略会影响个人的幸福感，但是对于时间碎片化使用的讨论非常有限。随着时间碎片化不再是具体事务的特征，而早已成为建构人们日常生活模式的物质基础。  

&emsp;&emsp;本文以2014-2015年的英国时间使用调查（UKTUS）为基础，针对时间的碎片化使用与自我感知的快乐感和自我感知的焦虑感这三个变量的关系进行分析，以期探讨时间的碎片化使用是否会影响个体自我认知的心理状态。

&emsp;&emsp;自我感知的快乐程度与焦虑程度均来源于原始的问卷数据，问卷要求用户自我报告他们近期的快乐和焦虑程度。  
&emsp;&emsp;在时间碎片化的指标上，我们参考了学者Gilbert Ritschard的操作化定义，拟用香农熵的值衡量碎片化程度。根据Shannon，熵是一种从信息科学中借用的统计度量(Shannon 1948)。Gilbert Ritschard提到，“在统计学中，熵反映了结果的不确定性或不可预测性。可能结果的多样性越高，不确定性就越高。香农熵是分类结果中最常用的多样性度量方法。通常，熵应用于离散分布，度量序列中状态的多样性，而状态分布可以看作是不同状态之间的时间分布。”

&emsp;&emsp;本文根据原始数据的结果，提出了两个反应幸福感的指标，分别是自我感知的快乐程度与自我感知的焦虑程度。并提出了以下假设：  
&emsp;&emsp;H1：时间使用碎片化程度越高，自我感知的快乐程度越低  
&emsp;&emsp;H2：时间使用碎片化程度越高，自我感知的焦虑程度越高

- &emsp;___状态种类数___

&emsp;&emsp;求一个个体在144个时间段内共有几个不同的状态

```python
#求一个个体在时间段内的状态总数 144个状态内一共有几个状态

def num_of_types(x):
    a = [0]*37
    for i in range(0,144):
        col = 'pri'+ str(i+1)
        for j in range(37):
            if x[col] == 100+1+j :
                a[j] = 1
    sum = 0
    for i in range (len(a)):
        sum = sum+a[i]
    return sum
```

```python
data.loc[:,"num_of_types"] = data.apply(num_of_types,axis=1)
```

- &emsp;___四项基本状态的种类数___

&emsp;&emsp;求个体的四项基本状态中占有几项  
&emsp;&emsp;a1-4分别代表四项活动：睡眠/休息、吃/喝、个人护理、旅行

```python
#求个体的四项基本状态中占有几项 a1-4分别代表四项活动

def num_of_basic(x):
    a = [0]*4
    for i in range(0,144):
        col = 'pri'+ str(i+1)
        #睡眠/休息
        if (x[col] == 101) | (x[col] == 102) :
            a[0] = 1
        #吃/喝
        if x[col] == 104 :
            a[1] = 1
        #个人护理
        if x[col] == 103 :
            a[3] = 1
        #旅行
        for j in range(112,117):
            if x[col] == j:
                a[3] = 1
        if (x[col] == 111) | (x[col] == 129) | (x[col] == 131):
            a[3] = 1

    sum = 0
    for i in range (len(a)):
        sum = sum+a[i]
    return sum
```

```python
data.loc[:,"num_of_basic"] = data.apply(num_of_basic,axis=1)
```

- &emsp;___个体睡眠时长___

&emsp;&emsp;求一个个体总计的睡眠时长

```python
#求个体的睡眠时长

def sleep_time(x):
    sleep = 0
    for i in range(0,144):
        col = 'pri'+ str(i+1)
        if x[col] == 101 :
            sleep = sleep+10
    return sleep
```

```python
data.loc[:,"sleep_time"] = data.apply(sleep_time,axis=1)
```

### &emsp;&emsp;3.3 变量处理

- &emsp;___时间使用碎片化——香农熵___

&emsp;&emsp;上文已经提到学者Gilbert Ritschard对碎片化的操作化定义，计算公式如下：

![example](https://wx4.sinaimg.cn/mw2000/006ohWzZly1gyr06dcen2j30ey03sdg9.jpg)

```python
import numpy as np
import pandas as pd

def entropy(x):
    a = [0]*38
    for i in range(0,144):
        col = 'pri'+ str(i+1)
        for j in range(38):
            if x[col] == 100+1+j :
                a[j] = a[j]+1
    entropy = 0
    for i in range(len(a)):
        if a[i] == 0:
            continue
        else:
            p = a[i]/144
            entropy += np.log2(p) * p * (-1)
    return entropy
```

```python
df.loc[:,"entropy"] = df.apply(entropy,axis=1)
```

```python
df['entropy'].value_counts()
```

## 五、讨论

### 1.时间使用碎片化与特定群体的关系

&emsp;&emsp;研究发现，时间使用碎片化具有群体性特征。

&emsp;&emsp;在性别维度，研究发现女性的时间使用碎片化程度较男性更高，由于女性在传统意义上一般承担较重的家务角色，所需关心的工作和生活事务也更为纷繁，故而更有碎片化的可能。  
&emsp;&emsp;在年龄维度，老年群体的时间使用碎片化程度较其他年龄段群体的程度更高，这是因为青少年及中年群体受制于教育、工作等因素，可能会在一个比较长的时间段内被要求从事同一件事，如上课等，而老年群体大多赋闲在家，有更充分的支配时间使用的便利，故而碎片化程度更高。  
&emsp;&emsp;在家庭人口维度，家庭人口越多则家庭事务也更繁重，且由于往往同处同一生活空间下，彼此生活事件发生地点较为集中，故而有更大可能在进行某一项活动时遭到打断或者需要同时进行多种事务以应付所需。在电子设备数量维度，越多的电子设备表明新技术的应用更为广泛，对传统生活的变革越大，对人们进行一项事务的帮助也越大，从而协助降低了人们对某一事物的关注程度而有更大可能在单位时间内处理更多事情。


### 2.时间使用碎片化与自我幸福感、焦虑感的关系

&emsp;&emsp;研究发现，时间使用碎片化程度越高，自我幸福感越低，焦虑感则越高。随着社会节奏的加快和信息海量程度的加剧，时间使用碎片化导致人们无法更好地沉浸在手头进行的事务中，同时对事物的耐心程度也随之下降，在海量的信息面前，人们变得越来越浮躁，注意力越来越低，越来越停留在浅层次的表达上，难以进入上手状态，从而可以获得的深层次理解和体验也越少，精神层面的依归和获得感也较低，从而导致个体往往感到心理上的空虚，幸福感也随之下降，焦虑程度则随之加深。


### 3.人口学变量与自我幸福感、焦虑感的关系

&emsp;&emsp;在人口学变量中，年龄和家庭人口数对自我幸福感和焦虑感的影响较为显著。在年龄维度，研究发现年龄越大，自我幸福感越高，焦虑感越低，这可能是因为随着人生阅历增长，逐渐对生活中的起伏变得淡然，所追求的目标也日益减少，偏重安稳，或是已经取得了想要的成果，安于生活，也有年长一辈所成长环境不同，追求的物质和符号需要可能并没有那么多或需求那么强烈。在家庭人口数维度，研究发现家庭人口数越多，幸福感越高，同时焦虑感也越高，这一方面体现了家庭兴旺下亲密关系数量增加对人心理的正向促进作用，一方面也表现出对现实生活中至亲之人事务的关切和担忧，情感联结越紧密，所需要考虑的事情和需承担的责任也随之增长。

&emsp;&emsp;而受教育程度与自我幸福感和焦虑感的相关性却不一。研究发现受教育程度越高，自我幸福感越高，但与自我焦虑感之间则无明显相关。受教育程度越高，个人的学识和社会地位也往往越高，获得感、尊严感，自我价值实现的程度也更深，前景也更为广阔，这或许是其自我幸福感较高的原因。而受教育程度的高低和焦虑之间也并无明显相关，或是因为不论受教育之多寡，人们总要面对现实生活中的一些共通的压力，特定生活环境中引起焦虑的维度也非常广泛，个体与个体之间的差异性也较大，无法简单地从单一维度进行解释，且教育年限的增长并不能解决诸多现实问题，还会在某些层面带来难以匹配对口、选择和改变成本更大的问题。



## 六、总结与反思

&emsp;&emsp;本研究对英国的时间使用数据进行了清洗与分析，在数据中计算个体报道的时间使用序列的香农熵来代表时间使用碎片化，并采用了问卷中的两个变量——自我报告的快乐程度与焦虑程度来窥探碎片化对个体心理的影响，将人口统计学变量、电子设备数量、时间是用碎片化程度、自我报告的快乐与焦虑程度为变量进行回归分析。研究发现，时间使用的碎片化程度越高，个体自我幸福感越低，焦虑感越高的显著关系，具有一定的学术价值与现实意义。

&emsp;&emsp;但不足之处在于，在研究方法上，本研究采用了横截面的方式，仅采用了时间使用调查中2014-2015一年中的样本，本研究的数据来源较为单一陈旧，不能很好地证明因果性；在内部效度上，变量的操作化定义——自我报告的快乐程度和焦虑程度仅采用了一个数据进行测量，不具有很好的内在效度，不能很好地代表用户的心理特征；在外部效度上，本研究是英国的时间数据，缺乏对不同个体所处文化背景和社会环境的分类分析；另外，研究的变量较少，缺乏对时间使用的碎片化具体的影响机制维度的揭示，尚需后来的研究者进一步予以揭示。
